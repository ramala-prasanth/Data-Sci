#Hypothesis Testing

```python
import numpy as np
import scipy.stats as st
```

#z-Test


```python
#H0 : μ = 46, Ha :  μ > 46
n = 30
xbar = 48.6
mu = 46
sigma = 4.2
alpha = 0.05
```

```python
z_critical = abs(st.norm.ppf(alpha)) #Absolute value taken as the it's a right-tailed test and the original value will be negative
z_critical
```

```python
z = (xbar-mu)/(sigma/np.sqrt(n))
z
```

```python
if (z < z_critical): #Right-tailed test
    print("Null hypothesis cannot be rejected")
else:
    print("Reject null hypothesis")
    
```

#z-Test using P-value

```python
#H0 : μ = 8 and Ha : μ != 8
n = 32
xbar = 8.2
mu = 8
sigma = 0.6
alpha = 0.05
```

```python
z = (xbar-mu)/(sigma/np.sqrt(n))
z
```

```python
p_val = (1 - st.norm.cdf(abs(z))) * 2
p_val
```

```python
if (p_val > alpha):
    print("Null hypothesis cannot be rejected")
else:
    print("Reject null hypothesis")
```

#T test


```python
#H0 : μ = 16.3, Ha : μ != 16.3
n = 10
degrees_of_freedom = n-1
xbar = 17.7
mu = 16.3
s = 1.8
alpha = 0.05
```

```python
t = (xbar - mu)/(s / np.sqrt(n))
t
```

```python
t_critical = st.t.ppf(alpha/2, degrees_of_freedom)
t_critical
```

```python
if (abs(t) > abs(t_critical)):
    print("Null hypothesis cannot be rejected")
else:
    print("Reject null hypothesis")
```

#t-Test using P-value

```python
#H0 : μ = 36.7, Ha : μ > 36.7
n = 15
degrees_of_freedom = n-1
xbar = 40.6
mu = 36.7
s = 6
alpha = 0.05
```

```python
t = (xbar - mu)/(s / np.sqrt(n))
t
```

```python
p_val = (1 - st.t.cdf(abs(t), degrees_of_freedom)) 
p_val
```

```python
if (p_val > alpha):
    print("Null hypothesis cannot be rejected")
else:
    print("Reject null hypothesis")
```
